# 投影変換

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def warpAffine():
    img = cv2.imread('images/sample4.png')
    
    pts1 = np.float32([[50,50],[200,50],[50,200]])
    pts2 = np.float32([[70,70],[200,50],[50,200]])
    
    M = cv2.getAffineTransform(pts1, pts2)
    
    result = cv2.warpAffine(img, M, (1000, 550))
    
    fig, axs = plt.subplots(nrows=1, ncols=2)
    axs[0].set_title('original')
    axs[0].imshow(img)
    axs[1].set_title('Affine Transform')
    axs[1].imshow(result)
    plt.show()
```

```python
warpAffine()
```
